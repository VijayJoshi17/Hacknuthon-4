```python
import smtplib
import ssl
import csv

# Email account details
EMAIL_ADDRESS = '21bce095@nirmauni.ac.in'
EMAIL_PASSWORD = '1234'

# Email details
subject = 'Your subject here'
body = 'Your email body here'

# Open the CSV file and read the data
with open('user_data.csv', 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    headers = next(csvreader)  # skip the header row
    data = [row for row in csvreader]

# Loop through the data and send an email to all users whose email_status is 'not sent'
for row in data:
    if row[3] == 'not sent':  # check email_status
        recipient = row[2]  # email address
        message = f'Subject: {subject}\n\n{body}'

        # Connect to the SMTP server and send the email
        context = ssl.create_default_context()
        with smtplib.SMTP_SSL('smtp.gmail.com', 465, context=context) as server:
            server.login(EMAIL_ADDRESS, EMAIL_PASSWORD)
            server.sendmail(EMAIL_ADDRESS, recipient, message)

        # Update the email_status for this user in the data
        row[3] = 'sent'

# Write the modified data back to the CSV file
with open('user_data.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(headers)
    csvwriter.writerows(data)
```
